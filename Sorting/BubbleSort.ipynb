```python
from graph import draw_graph
import pandas as pd
```

```python
# function to perform bubble sort
def bubble_sort(arr, sorttype="ascending"):
    n = len(arr)
    df = pd.DataFrame()
    
    df['number of arguements'] = list(range(0, n))
    df['values'] = arr
    
    draw_graph(x='number of arguements', y='values', data=df, title='Original input')
    
    step = 0
    for i in range(0, n):
        step = step + 1
        
        for j in range(0, n-1-i):
            if sorttype == "descending":
                if arr[j] < arr[j+1]:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
            else:
                if arr[j] > arr[j+1]:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
        
        df['values'] = arr
        draw_graph(x='number of arguements', y='values', data=df, title="Step "+str(step))
        
```

```python
arr = [5, 1, 15, 22, 21, 0, 3, 6, 9, 2]
bubble_sort(arr)
```
