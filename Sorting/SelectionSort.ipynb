```python
# importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# function to visualize the step by step state of the array
def draw_graph(x, y, data, title):
    sns.set(style="darkgrid", rc={'figure.figsize':(5, 3)})
    
    ax = sns.barplot(x=x, y=y, data=data)
    ax.set_title(title, fontsize=20)
    
    plt.show()
```

```python
# function to perform selection sort
def selection_sort(arr, sorttype="ascending"):
    n = len(arr)
    df = pd.DataFrame()
    
    df['number of arguements'] = list(range(0, n))
    df['values'] = arr
    
    draw_graph(x='number of arguements', y='values', data=df, title='Original input')
    
    step = 0
    for i in range(0, n):
        step = step + 1
        
        for j in range(i+1, n):
            if sorttype == "descending":
                if arr[i] < arr[j]:
                    arr[i], arr[j] = arr[j], arr[i]
            else:
                if arr[i] > arr[j]:
                    arr[i], arr[j] = arr[j], arr[i]
        
        df['values'] = arr
        draw_graph(x='number of arguements', y='values', data=df, title="Step "+str(step))
        
```

```python
arr = [12, 1, 3, 8, 0, 20, 15, 11]
selection_sort(arr)
```
