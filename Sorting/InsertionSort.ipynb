```python
# importing libraries
from graph import draw_graph
import pandas as pd
```

```python
# function to perform insertion sort
def insertion_sort(arr, sorttype='ascending'):
    n = len(arr)
    df = pd.DataFrame()
    
    df['number of arguements'] = list(range(0, n))
    df['values'] = arr
    
    draw_graph(x='number of arguements', y='values', data=df, title='Original input')
    
    step = 0
    for i in range(1, n):
        key = arr[i]
        j = i-1
        step = step + 1
        
        if sorttype == 'descending':
            while j>=0 and key>arr[j]:
                arr[j+1] = arr[j]
                j = j-1
            arr[j+1] = key
        else:
            while j>=0 and key<arr[j]:
                arr[j+1] = arr[j]
                j = j-1
            arr[j+1] = key
        
        df['values'] = arr
        draw_graph(x='number of arguements', y='values', data=df, title="Step "+str(step))
```

```python
arr = [5, 2, 1, 10, 11, 3, 6, 4]
insertion_sort(arr, sorttype='descending')
```
